###Data Logging

```python
from datetime import datetime

def log(message):
  timestamp_format = '%d-%h-%Y-%H:%M:%S'
  now = datetime.now()
  timestamp = now.strftime(timestamp_format)
  with open("logfile.txt", "a") as f:
    f.write(timestamp + ',' + message + '\n')
```

```python
log("ETL Job Started")
```

```python
import glob
import pandas as pd
import numpy as np
from natsort import index_natsorted
```

```python
from posixpath import sep
def extract_from_json(file_to_process):
    dataframe = pd.read_json(file_to_process,lines=True)
    return dataframe

def extract_from_csv(file_to_process): 
    dataframe = pd.read_csv(file_to_process) 
    return dataframe

```

```python
extract_from_csv('station-information.csv')
```

```python
extract_from_json('station-information.json')
```

```python
def extract():
    extracted_data = pd.DataFrame(columns=['station_id','name','region_id', 'lon', 'lat', 'address'])
    #for csv files
    for csvfile in glob.glob("*.csv"):
      extracted_data = extracted_data.append(extract_from_csv('station-information.csv'), ignore_index=True) 
    #for json files
    for jsonfile in glob.glob("*.json"):
      extracted_data = extracted_data.append(extract_from_json('station-information.json'), ignore_index=True)
    return extracted_data
```

```python
df_info = extract()
df_info
```

```python
df_bike = extract_from_json('bike.json')
df_bike
```

```python
df_region = extract_from_json('system-regions.json')
df_region
```

```python
df_status = extract_from_json('station-status.json')

df_status
```

##Transform

```python
log("Transform phase Started")
```

###Dataset Stasiun

```python
df_info['station_id'] = df_info['station_id'].str.replace(r'\D', '')
df_info
```

```python
df_info['region_id'] = df_info['region_id'].str.replace(r'\D', '')
df_info
```

```python
duplicate = df_info[df_info.duplicated()]
duplicate
```

```python
df_info = df_info.drop_duplicates(ignore_index=True)
```

```python
df_info
```

```python
df_info=df_info.sort_values(
   by="station_id",
   key=lambda x: np.argsort(index_natsorted(df_info["station_id"])), ignore_index=True
)
```

```python
df_info
```

```python
df_info = df_info.drop(columns = ['lon', 'lat'])
```

```python
df_info = df_info.rename({'name': 'station_name'}, axis=1)

df_info
```

###Dataset Sepeda

```python
df_bike['bike_id'] = df_bike['bike_id'].str.replace(r'\D', '')
df_bike
```

```python
df_bike['region_id'] = df_bike['region_id'].str.replace(r'\D', '')
df_bike
```

```python
df_bike = df_bike.rename({'name':'bike_name', 'region_id':'region_id'}, axis=1)  # new method

df_bike
```

```python
df_bike = df_bike.drop(columns = ['is_reserved', 'is_disabled'])
```

```python
df_bike
```

```python
df_bike = df_bike.sort_values(
   by="bike_id",
   key=lambda x: np.argsort(index_natsorted(df_bike["bike_id"])), ignore_index=True
)
```

```python
df_bike
```

###Dataset Negara

```python
df_region['region_id'] = df_bike['region_id'].str.replace(r'\D', '')
df_region
```
